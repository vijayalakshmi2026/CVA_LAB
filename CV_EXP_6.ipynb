```python
import cv2
import time
from google.colab.patches import cv2_imshow  # Import cv2_imshow for Colab

# Load pre-trained face detection model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the input image
image_path = "/content/sample_data/download (1).jpg"  # Ensure you uploaded the correct image
image = cv2.imread("/content/sample_data/download (1).jpg")

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces
start_time = time.time()
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))
end_time = time.time()

# Define labels for detected faces
names = ["Alex", "Jessi"]  # Modify names as needed

# Draw bounding boxes and labels
for i, (x, y, w, h) in enumerate(faces):
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    label = names[i] if i < len(names) else "Person"
    cv2.putText(image, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)


# Show the output in Colab
cv2_imshow(image)  # ✅ Corrected: Only one argument (image)
```
